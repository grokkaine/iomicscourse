## Network fusion


- think:
	- can you apply this method in your project, under what conditions?
	- how would you prepare the data, and compute similarities?
	- how would you use the results of network fusion in your project?
	- how would you fit parameters?
	- how would you use this method for feature transformation?
	- what clustering strategies could you take?


- practical:
	- use one of the supplied network fusion libraries
	- compute similarity graphs from different omics levels
	- attempt a hyper-parametrization study
	- perform clustering on a similarity network (single omics or fused)
	- perform survival analisys, pathway enrichment etc
	- try consensus clustering or spectral clustering on your data


- coding:
	- code a NF method yourself, in a language of your choice
	- atempt to generate a synthetic dataset based on a fused model
	- (for img processing skills) perform NF on two images


## Matrix factorization and reccomender systems


- think:
	- how to use omics datasets from a reccommender systems perspective
	- what other MF methods besides NMF can you use? Example: can you use autoencoders? How?
	- Could an autoencoder be used to compute NMF?
	- is your project data good to be considered for an NMF method?
	- is your project interesting from a reccomender systems perspective?
	- could you transform your data to be useful in a MF/NMF setting?
	- learn to generate synthetic datasets based on hidden factors


- practical:
	- start with the result factor matrices, how do you compute clusters?
	- use one of the supplied NMF libraries
	- how do you follow up the results
	- what can be parametrized, how to fit rank estimates?
	- attempt different MF methods (SVD, PCA) and compare.
	- compare clustering results from different methods (kmeans, nmf, etc)


- code:
	- code your own ALS or SGD method in a MF settings
	- generate synthetic data, in a supervised or unsupervised setting
	- (for image processing fanatics) extract hidden features via NMF and compare to autoencoders
	- (for NN gurus) code a deep NMF
